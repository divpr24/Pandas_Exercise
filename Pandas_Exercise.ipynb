```python
#Example Questions:
#1.
import pandas as pd
temp_celcius = [34,30,40,38,27]
city = ["Delhi","Mumbai","Chennai","Kolkata","Ranchi"]

series_1 = pd.Series(temp_celcius, index = ["City1","City2","City3","City4","City5"])
series_2 = pd.Series(city, index = ["City1","City2","City3","City4","City5"])

print(series_1)
print("\n")
print(series_2)
print("\n")
print(series_1.sort_values())
print("\n")
print(series_2.filter(items = ["City1", "City2"]))
```
